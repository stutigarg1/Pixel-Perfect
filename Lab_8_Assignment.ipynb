```python
import numpy as np
import matplotlib.pyplot as plt

# Create a noisy 1-D signal
x = np.linspace(0, 10, 100)
signal = np.sin(x) + np.random.normal(0, 0.5, x.shape)

# Apply 1-D convolution with a moving average filter
filtered_signal = np.convolve(signal, np.ones(5)/5, mode='same')

# Plot
plt.plot(x, signal, label='Noisy Signal')
plt.plot(x, filtered_signal, label='Filtered Signal', linewidth=2)
plt.title('1-D Convolution for Noise Removal')
plt.legend()
plt.show()
```

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# Load image and add noise
image = cv2.imread('/content/flower.PNG', cv2.IMREAD_GRAYSCALE)  # Replace with your image path
noisy_image = np.clip(image + np.random.normal(0, 25, image.shape), 0, 255).astype(np.uint8)

# Apply Gaussian blur for noise reduction
filtered_image = cv2.GaussianBlur(noisy_image, (5, 5), 1)

# Resize images
new_size = (300, 300)  # Specify new size
noisy_image_resized = cv2.resize(noisy_image, new_size)
filtered_image_resized = cv2.resize(filtered_image, new_size)

# Display images using cv2_imshow
print('Noisy Image:')
cv2_imshow(noisy_image_resized)
print('Filtered Image:')
cv2_imshow(filtered_image_resized)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread('/content/flower.PNG', cv2.IMREAD_GRAYSCALE)

# Define the Low Pass Filter kernel
lpf_kernel = np.array([[1/9, 1/9, 1/9],
                       [1/9, 1/9, 1/9],
                       [1/9, 1/9, 1/9]])

# Apply Low Pass Filter
lpf_image = cv2.filter2D(image, -1, lpf_kernel)

# Display the Low Pass Filtered Image
plt.imshow(lpf_image, cmap='gray')
plt.title('Low Pass Filtered Image')
plt.axis('off')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread('/content/flower.PNG', cv2.IMREAD_GRAYSCALE)

# Define the High Pass Filter kernel
hpf_kernel = np.array([[0, -1, 0],
                       [-1, 5, -1],
                       [0, -1, 0]])

# Apply High Pass Filter
hpf_image = cv2.filter2D(image, -1, hpf_kernel)

# Display the High Pass Filtered Image
plt.imshow(hpf_image, cmap='gray')
plt.title('High Pass Filtered Image')
plt.axis('off')
plt.show()
```
