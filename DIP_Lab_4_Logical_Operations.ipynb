```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
from google.colab import files

# Upload images
uploaded = files.upload()
```

```python
# Converting the images in grayscale
valley = cv2.imread('valley.jpg', cv2.IMREAD_GRAYSCALE)
theo = cv2.imread('theo.jpg', cv2.IMREAD_GRAYSCALE)

# Resize 'theo' if the shapes do not match
if valley.shape != theo.shape:
    theo = cv2.resize(theo, (valley.shape[1], valley.shape[0]))
```

```python
# AND operation
and_result = cv2.bitwise_and(valley, theo)

# Display the original images and the result
plt.figure(figsize=(10, 10))

plt.subplot(1, 3, 1)
plt.imshow(valley, cmap='gray')
plt.title('Valley Image')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(theo, cmap='gray')
plt.title('Theo Image')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(and_result, cmap='gray')
plt.title('AND Operation Result')
plt.axis('off')

plt.show()
```

```python
# OR operation
or_result = cv2.bitwise_or(valley, theo)

# Display the original images and the result
plt.figure(figsize=(10, 10))

plt.subplot(1, 3, 1)
plt.imshow(valley, cmap='gray')
plt.title('Valley Image')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(theo, cmap='gray')
plt.title('Theo Image')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(or_result, cmap='gray')
plt.title('OR Operation Result')
plt.axis('off')

plt.show()
```

```python
#Intersection
# Thresholding to create binary images
_, valley_binary = cv2.threshold(valley, 1, 255, cv2.THRESH_BINARY)
_, theo_binary = cv2.threshold(theo, 1, 255, cv2.THRESH_BINARY)

# Perform the AND operation on the binary images to find the intersection
intersection_result = cv2.bitwise_and(valley_binary, theo_binary)

# Display the original images and the intersection result
plt.figure(figsize=(10, 10))

plt.subplot(1, 3, 1)
plt.imshow(valley, cmap='gray')
plt.title('Valley Image')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(theo, cmap='gray')
plt.title('Theo Image')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(intersection_result, cmap='gray')
plt.title('Intersection Result')
plt.axis('off')

plt.show()
```

```python
# Watermarking-- XOR operation
xor_result = cv2.bitwise_xor(valley, theo)

# Display the original images and the XOR result
plt.figure(figsize=(10, 10))

plt.subplot(1, 3, 1)
plt.imshow(valley, cmap='gray')
plt.title('Image 1')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(theo, cmap='gray')
plt.title('Image 2')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(xor_result, cmap='gray')
plt.title('XOR Result')
plt.axis('off')

plt.show()
```

```python
# NOT operation
not_valley = cv2.bitwise_not(valley)
not_theo = cv2.bitwise_not(theo)

# Display the original images and the NOT results
plt.figure(figsize=(10, 8))

# Original Valley Image
plt.subplot(2, 2, 1)
plt.imshow(valley, cmap='gray')
plt.title('Valley Image')
plt.axis('off')

# NOT Valley Image
plt.subplot(2, 2, 2)
plt.imshow(not_valley, cmap='gray')
plt.title('NOT Valley Image')
plt.axis('off')

# Original Theo Image
plt.subplot(2, 2, 3)
plt.imshow(theo, cmap='gray')
plt.title('Theo Image')
plt.axis('off')

# NOT Theo Image
plt.subplot(2, 2, 4)
plt.imshow(not_theo, cmap='gray')
plt.title('NOT Theo Image')
plt.axis('off')

plt.tight_layout()  # Adjust subplots to fit in figure area.
plt.show()
```
