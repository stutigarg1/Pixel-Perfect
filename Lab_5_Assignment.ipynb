```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
from google.colab.patches import cv2_imshow
```

```python
image=cv2.imread('/content/tiger.png')
```

```python
scaled_image = cv2.resize(image, None, fx=0.5, fy=0.5, interpolation=cv2.INTER_LINEAR)
cv2_imshow(scaled_image)
print('Scaled Image')
```

```python
rows, cols = image.shape[:2]
Rotation_matrix = cv2.getRotationMatrix2D((cols/2, rows/2), 45, 1)
rotated_image = cv2.warpAffine(image, Rotation_matrix, (cols, rows))
cv2_imshow(rotated_image)
```

```python
SHRINK_FACTOR=0.5
shrunk_image = cv2.resize(image, None, fx=0.5, fy=0.5, interpolation=cv2.INTER_AREA)
cv2_imshow(shrunk_image)
print('Shrunk Image')
```

```python
ZOOM_image = cv2.resize(image, None, fx=2, fy=2, interpolation=cv2.INTER_CUBIC)
cv2_imshow(ZOOM_image)
print('ZOOM Image')
```

```python
TRANSLATION_MATRIX = np.float32([[1, 0, 100], [0, 1, 100]])
TRANSLATED_IMAGE = cv2.warpAffine(image, TRANSLATION_MATRIX, (cols, rows))
cv2_imshow(TRANSLATED_IMAGE)
```
